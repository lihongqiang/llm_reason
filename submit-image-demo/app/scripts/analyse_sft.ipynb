```python
import pandas as pd

# qwen2
model = 'Qwen2-7B-Instruct'
start_prefix_list = ['Assistant: ', 'answer\n', 'assistant\n', 'user\n', 'output\n', 'kle: ', 'doc', '_response\n', ' Assistant\n', 'docstream\n', 'rust\n', 'document\n', 'step\n', 'answers:\n', ' actionable steps to solve the problem:\n', 'adicus\n', 'url\n', 'step\n', '用户\n', 'year\n\n', 'classifications\n', 'answer:\n', 'refined_data\n', 'psilon\n\n', ' Germans\n', '崽回答：\n\n\n', 'ease\n', '唯美渐变背景\n\n        \n       ', 'docstring\n', ' Premise:\n', 'analyze\n\n', '_RESOLUTION:\n', ' Cement the rationale:\n', 'author\n', '(answer)\n', 'jenis:\n', ' reasoning:\n\n', ' Poloish the question:\n\n        ', ' osgtidq 北京\n\n        ', 'step response:\n']

# llama3.1
# model = 'Meta-Llama-3.1-8B-Instruct'
# start_prefix_list = ['<|start_header_id|>assistant<|end_header_id|>\n\n']

# glm4
# model = 'glm-4-9b-chat'
# start_prefix_list = []

data = pd.read_json(f'/data/root/submit-image-demo/app/data/{model}_sft.jsonl')
data.head()
```

```python
# 去掉格式
for start_prefix in start_prefix_list:
    data['output'] = data['output'].apply(lambda x: x.replace(start_prefix, ''))
data.head()
```
