```python
import os
import json
import pandas as pd
data_file = 'Meta-Llama-3.1-405B-Instruct-FP8_sft.jsonl'
if os.path.exists(data_file):
    with open(data_file, 'r') as f:
        lines = f.readlines()
        data = json.loads(' '.join(["["]+lines+["]"]))
data = pd.DataFrame(data)
prefix = '\n\n        ### 答案:\n'
data['instruction'] = data['instruction'].apply(lambda x: x[:x.find(prefix)])
prefix = '<|start_header_id|>assistant<|end_header_id|>\n\n'
data['output'] = data['output'].apply(lambda x: x.replace(prefix, ""))

data.head()
```

```python
data.to_json('Meta-Llama-3.1-405B-Instruct-FP8_sft_fix.jsonl', orient='records', indent=4, force_ascii=False)
```
